```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# 1. Filter schools meeting the 80% math score threshold (640/800)
best_math_schools = schools.loc[schools['average_math'] >= (800 * 0.8), ['school_name', 'average_math']].sort_values('average_math', ascending=False)    
    
best_math_schools
```

```python
import plotly.express as px
# Shorten labels by removing redundant 'High School' suffix
best_math_schools["school_name"] = best_math_schools["school_name"].str.replace("High School", "", regex=False).str.strip()
# Sort ascending so highest score appears at the top of the horizontal bar chart
best_math_schools = best_math_schools.sort_values("average_math", ascending=True)

fig = px.bar(best_math_schools,
             x="average_math",
             y="school_name",
             orientation="h",
             title="Best Math Schools")

fig.update_layout(
    xaxis_title="Average Math Score",
    yaxis_title="School",
    xaxis=dict(showline=True, linecolor="black"),
    yaxis=dict(showline=True, linecolor="black")
)
# Display score values inside each bar
fig.update_traces(texttemplate="%{x}", textposition="inside")
fig.show()
fig.write_image("best_math_schools.png")
```

```python
# 2. Compute total SAT score as the sum of all three sections
schools['total_SAT'] = schools['average_math'] + schools['average_reading'] + schools['average_writing']
# Select and rank the top 10 schools by total SAT score
top_10_schools = schools[['school_name', 'total_SAT']].sort_values('total_SAT', ascending=False).head(10)

top_10_schools 
```

```python
import plotly.express as px
# Shorten labels by removing redundant 'High School' suffix
top_10_schools ["school_name"] = top_10_schools ["school_name"].str.replace("High School", "", regex=False).str.strip()
# Sort ascending so highest score appears at the top of the horizontal bar chart
top_10_schools  = top_10_schools .sort_values("total_SAT", ascending=True)

fig = px.bar(top_10_schools,
             x="total_SAT",
             y="school_name",
             orientation="h",
             title="Top 10 Schools")

fig.update_layout(
    xaxis_title="Total SAT",
    yaxis_title="School",
    xaxis=dict(showline=True, linecolor="black"),
    yaxis=dict(showline=True, linecolor="black")
)
# Display score values inside each bar
fig.update_traces(texttemplate="%{x}", textposition="inside")
fig.show()
fig.write_image("top_10_schools.png")
```

```python
# The single borough with the largest standard deviation in the combined SAT score
largest_std_dev =  schools.groupby('borough')

largest_std_dev = largest_std_dev.agg(
    std_SAT=('total_SAT', 'std'),        # Standard deviation of total SAT scores
    num_schools=('total_SAT', 'count'),  # Number of schools per borough
    average_SAT=('total_SAT', 'mean')    # Mean total SAT score per borough
).reset_index()
# Round results and isolate the borough with the highest variability
largest_std_dev = largest_std_dev.round(2).nlargest(1, 'std_SAT')

largest_std_dev
```

```python
# Recompute full borough stats for comparative visualization
borough_stats =  schools.groupby('borough')

borough_stats = borough_stats.agg(
    std_SAT=('total_SAT', 'std'),        # Standard deviation of total SAT scores
    num_schools=('total_SAT', 'count'),  # Number of schools per borough
    average_SAT=('total_SAT', 'mean') 
).reset_index()

borough_stats = borough_stats.round(2)
borough_stats
```

```python
import plotly.express as px

# Sort boroughs by standard deviation for visual clarity
borough_stats  = borough_stats.sort_values("std_SAT", ascending=True)

fig = px.bar(borough_stats,
             x="borough",
             y="std_SAT",
             title="Boroughs With The Standard Deviation Of The Total SAT")

fig.update_layout(
    xaxis_title="Borough",
    yaxis_title="Std SAT",
    xaxis=dict(showline=True, linecolor="black"),
    yaxis=dict(showline=True, linecolor="black")
)
# Display standard deviation values inside each bar
fig.update_traces(texttemplate="%{y}", textposition="inside")
fig.show()
fig.write_image("borough_std_dev.png")
```
